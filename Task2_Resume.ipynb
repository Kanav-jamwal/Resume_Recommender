### Series of steps:
Now we got the list of resumes on which we are going to perform various steps for pre-processing data:
    
1. Word tokenizer(\w+(?:[-']\w+)?) given as per the assignment specification.
2. Sentence tokenizer-This tokenizer divides a text into a list of sentences
by using an unsupervised algorithm to build a model for abbreviation
words, collocations, and words that start sentences.  It must be
trained on a large collection of plaintext in the target language
before it can be used.(nltk organization)
3. Using strip to remove \n character.                        
4. Removing non-alpha character such as digit.
5. Removing words with word length less than 3.

Defining the tokenizer.

```python
tokenizer = RegexpTokenizer(r"\w+(?:[-']\w+)?")
```

Function for tokenizing the data.

```python
def tokenize(x):
    tokenized_data = tokenizer.tokenize(x)
    return(tokenized_data)
```

We are reading every file in final_resume and than appending the data into data_set.

```python
data_set={}
for each in final_resume:
    with open(each, 'r',encoding='utf8') as file:
        data=file.read()
        data_set[each] = data
        
```

#####  Steps performed on raw data:
1. Word Tokeizer
2. Sentence Tokenizer
3. Strip function to remove \n kind of character.
4. Making the first character of the sentence after applying the sentence segmentation checking the character of sentence and replacing it with lower().

#### Explaination
We have already loaded the data into data_set now we are using the sentence tokenizer and strip on the raw data to get a clean 
data for analysys.
After sentence tokenization we are going for the word tokeiztion and are replacing the first character of the sentence to lower() by using the tokenize(arg) for the word tokenizer and .lower() for replacement and in the appending the data into the dictionary..  

```python
data_set_tok = {}
```

```python
def sentence_break():
    sent_detector = nltk.data.load('tokenizers/punkt/english.pickle')
    for each in data_set:
        big_list = []
        sentences = sent_detector.tokenize(data_set[each].strip())
        for each1 in sentences:
            ret_value = []
            ret_value = tokenize(each1)
            if(len(ret_value) > 0):
                ret_value[0] = ret_value[0].lower()
            big_list += ret_value
        data_set_tok[each] = big_list

sentence_break()
```

### Removing non-alpha character

For the bigrams I think so we should remove the date format and the word as well which contains number in between as it states meaningful bigrams.Now removing non-alpha character from the dictionart and appending the data into a new dictionary.


```python
data_set_tok1 = {}
for each in data_set_tok:
    temp_list1 = []
    for each1 in data_set_tok[each]:
        if(each1.isalpha()):
            temp_list1.append(each1)
    data_set_tok1[each] = temp_list1
    
```

### Removing the word with word length greater than 3.

Using for loop and len for every element less than length 3 would be removed and than adding the words into the new dictionary.

```python
data_set_tok2 = {}
for each in data_set_tok1:
    temp_list1 = []
    for each1 in data_set_tok1[each]:
        if(len(each1)>2):
            temp_list1.append(each1)
    data_set_tok2[each] = temp_list1
    
    
```

Making the list of the latest dictionary which contains the updated vocabulary using iterable.

```python
final_words= list(chain.from_iterable(data_set_tok2.values()))
```

### Bigrams using nltk dictionary

We are finding bigrams using Bigram methods from nltk library in which we are matching the frequency to see how many times the bigrams are coming together in the data.
We have set the value of frequency filter to 18.

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
bigram_finder = nltk.collocations.BigramCollocationFinder.from_words(final_words)
bigram_finder.apply_freq_filter(18)
bigram_finder.apply_word_filter(lambda w: len(w) < 3)
bigrams_200 = bigram_finder.nbest(bigram_measures.pmi, 200)
bigrams_200
```

### Updating in dictionary.
Now, matching the values of bigrams in the dictionary and than updating the bigrams into vocab list. 

```python
#check found bigrams in my original data
mwetokenizer = MWETokenizer(bigrams_200)
token_dic_bigram =  dict((k, mwetokenizer.tokenize(v)) for k,v in data_set_tok2.items())
#words in all data
vocab = list(chain.from_iterable(token_dic_bigram.values()))
```

### Series of steps:
1. Using stopwords file provided in the assignment remove the stopword from the list of the vocab.
2. Removing context dependent stop words as stated in the assignment to remove the words which got more than 98% and less than    2% in the vocab list.
3. Using porter stemming stem the vocab list.
4. The last operation to sort out the vocab list.

Reading the stop word and storing it.

```python
stopword_file=open('stopwords_en.txt',"r",encoding="utf8")
stop_list=stopword_file.read().split('\n')
```

Removing the duplicates from stop and storing it in list.

```python
stop_list = list(set(stop_list))
```

Now, removing the stopwords from vocab list and storing it in new list.

```python
vocab1 = []
vocab1 = [w for w in vocab if w not in stop_list]
```

```python
vocab1 = list(set(vocab1))
```

### Removing context dependent stop word

By using the percentage required for removing the context dependent words we are eliminating the words that fall into that category.

```python
vocab2 = []
for each in vocab1:
    doc_count = 0
    for each1 in data_set_tok2:
        if(each in data_set_tok2[each1]):
            doc_count+=1
    if((doc_count > (len(final_resume)*2/100) and doc_count < (len(final_resume)*98/100) )):
        vocab2.append(each)
```

```python
len(vocab2)
```

### Stemming

Using the porter stemmer library of nltk we are performing stemming on the last updated vocab.
Taking code from tutorial as well.

```python
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
tagged_vocs = []
tagged_voc = nltk.tag.pos_tag(vocab2)
tagged_vocs.append(tagged_voc)
```

```python
stemmer = PorterStemmer()
vocab3 =[]
for tagged_set in tagged_vocs:
    vocab3 += [stemmer.stem(w[0]) for w in tagged_set ]
```

```python
vocab3 = list(set(vocab3))
```

### Check for vocab count

Checking for the tagged words and updating it in the dictionary.



```python
dic_stemmed = {}
for each in data_set_tok2:
    tagged_vocs = []
    tagged_voc = nltk.tag.pos_tag(data_set_tok2[each])
    tagged_vocs.append(tagged_voc)
    tokens_stemmer =[]

    for tagged_set in tagged_vocs:
        tokens_stemmer += [stemmer.stem(w[0]) for w in tagged_set ]
    dic_stemmed[each] = tokens_stemmer
```

```python
vocab_count = {}
for each in dic_stemmed:
    temp = {}
    for each1 in dic_stemmed[each]:
        if(each1 not in temp ):
            if(each1 in vocab3):
                temp[vocab3.index(each1)] = dic_stemmed[each].count(each1)
        vocab_count[each] = temp
        
```

### Output Word Count Vocab for each resume

```python
output_file=""
file = open("29554209_CountVec.txt", "w")
for each in vocab_count:
    output_file += each + ", "
    temp1 = vocab_count[each]
    for each1 in temp1:
        output_file += (str(each1) + ":" + str(temp1[each1])+ ", ")
    output_file +="\n"
file.write(output_file)
file.close()
```
